# Πόσο ελεύθεροι και ασφαλείς είναι άραγε οι δημοσιογράφοι στη Μέση Ανατολή;

#### Τα τελευταία χρόνια τα γεγονότα που επιβεβαιώνουν το έλλειμμα δημοκρατίας και ελευθερίας στις χώρες της Μέσης Ανατολής πληθαίνουν. Σε τέτοιες περιπτώσεις η ελεύθερη, ανεξάρτητη και ασφαλής δημοσιογραφία είναι πιο σημαντική από ποτέ. Τελικά, είναι όμως έτσι στην πράξη;

(photo)

Έχουν περάσει σχεδόν 3 μήνες από την έναρξη του νέου πολέμου μεταξύ Ισραήλ και Χαμάς και οι αριθμοί των θανάτων, των τραυματισμών και των υλικών καταστροφών ολοένα και αυξάνονται. Ανάμεσα σε αυτούς τους αριθμούς βρίσκεται και ο αριθμός των επιθέσεων εναντίον των δημοσιογράφων και εργαζομένων στα ΜΜΕ, ο οποίος σύμφωνα με τη Committee to Protect Journalists (CPJ) «οδήγησε στην πιο θανατηφόρα περίοδο για τους δημοσιογράφους από τότε που η CPJ άρχισε να συλλέγει δεδομένα το 1992». Μέχρι τις 4 Ιανουαρίου 2024, οι έρευνες της CPJ έδειξαν ότι τουλάχιστον 77 δημοσιογράφοι και εργαζόμενοι στα μέσα ενημέρωσης συγκαταλέγονταν στους περισσότερους από 23.000 νεκρούς από την έναρξη του πολέμου στις 7 Οκτωβρίου.

Είναι όμως μόνο οι επιθέσεις που καταγράφονται σε περίοδο πολέμου που εγείρουν ανησυχίες για την ελευθερία και την ασφάλεια των δημοσιογράφων στη Μέση Ανατολή; Όπως αποδεικνύουν οι έρευνες των Δημοσιογράφων Χωρίς Σύνορα, υπάρχουν και άλλες περιπτώσεις σε χώρες της Μέσης Ανατολής κατά τις οποίες ο Τύπος φιμώνεται με τον έναν ή τον άλλον τρόπο. Έξι χρόνια μετά την απόπειρα πραξικοπήματος, μετά την οποία η Τουρκία έγινε η μεγαλύτερη φυλακή δημοσιογράφων στον κόσμο, μέσα σε μόλις μισό χρόνο, 30 δημοσιογράφοι που εργάζονταν στα φιλοκουρδικά μέσα ενημέρωσης συνελήφθησαν και 25 από αυτούς φυλακίστηκαν (στοιχεία Reporters Without Borders), χωρίς καν να κατηγορηθούν ή να καταδικαστούν. Επίσης, στο Ιράν, έπειτα από το θάνατο της Mahsa Amini στις 16 Σεπτεμβρίου του 2022, ενώ τελούσε υπό κράτηση της αστυνομίας των ηθών, συνολικά 79 δημοσιογράφοι που προσπάθησαν να ερευνήσουν την υπόθεση, φυλακίστηκαν με την κατηγορία της αντικυβερνητικής προπαγάνδας (στοιχεία Reporters Without Borders).

Αυτές οι περιπτώσεις, όμως, δεν αποτελούν είδηση για τους δημοσιογράφους στη Μέση Ανατολή. Μόνο στον 21ο αιώνα η Μέση Ανατολή μετρά 25 διαφορετικές εμπόλεμες συρράξεις, πολλές από τις οποίες είναι ακόμα εν εξελίξει, ενώ οι πολιτικές και κοινωνικές αναταραχές θα μπορούσε να πει κανείς πως δεν έχουν ποτέ τελειωμό. Με τις χώρες της Μέσης Ανατολής να βρίσκονται μεταξύ των θέσεων 97 και 177 στον Παγκόσμιο Δείκτη Ελευθερίας του Τύπου, πόσο ελεύθεροι και ασφαλείς είναι άραγε οι δημοσιογράφοι στη Μέση Ανατολή;
