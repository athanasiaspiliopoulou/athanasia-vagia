## DOCUMENTATION

**Σε αυτό το αρχείο περιλαμβάνεται η επεξήγηση κάθε μεταβλητής στο major dataset που δημιουργήθηκε έπειτα από τη συλλογή των δεδεδομένων από την ιστοσελίδα της Committee to Protect Journalists με σκοπό να πραγματοποιηθεί το ρεπορτάζ με τίτλο "Πόσο ελεύθεροι και ασφαλείς είναι άραγε οι δημοσιογράφοι στη Μέση Ανατολή;" το οποίο υπογράφουν οι Αθανασία Σπηλιοπούλου και Βαϊα-Μαρία Τάτα.**

**Column Names:** "name", "gender", "organisation", "date", "location", "type_of_death", "attack", "role", "region", "year", "staff_freelancer"

1. **name** = το ονοματεπώνυμο κάθε δημοσιογράφου ή εργαζομένου στα ΜΜΕ που δέχθηκε την εκάστοτε επίθεση.
2. **gender** = το φύλο του κάθε δημοσιογράφου ή εργαζομένου στα ΜΜΕ που δέχθηκε την εκάστοτε επίθεση.
   - Στις περιπτώσεις δημοσιογράφων που έχουν σκοτωθεί το φύλο είναι γνωστό, όμως στις περιπτώσεις των υπόλοιπων εργαζομένων στα ΜΜΕ που έχουν σκοτωθεί η Committee to Protect Journalists δεν παρέχει αντίστοιχες πληροφορίες.
3. **organization** = το μέσο ή ο οργανισμός όπου εργάζεται ο εκάστοτε δημοσιογράφος ή εργαζόμενος στα ΜΜΕ ή αν εργάζεται ως freelancer.
4. **date** = η ημερομηνία που πραγματοποιήθηκε η εκάστοτε επίθεση.
5. **location** = η χώρα στην οποία πραγματοποιήθηκε η εκάστοτε επίθεση.
6. **type_of_death** = το κίνητρο ή η αιτία που οδήγησε στο θάνατο του εκάστοτε δημοσιογράφου, αν παρέχεται ως πληροφορία.
   - Στις περιπτώσεις θανάτων των εργαζομένων στα ΜΜΕ δεν διευκρινίζεται το κίνητρο που οδήγησε στους θανάτους τους, καθώς σύμφωνα με τη Committee to Protect Journalists, στην καταμέτρηση αυτών των επιθέσεων συγκαταλέγονται και οι εργαζόμενοι στα ΜΜΕ εκτός από τους δημοσιογράφους, ως ένδειξη σεβασμού για τη συμβολή τους στη δημοσιογραφία, ενώ έχασαν τη ζωή τους στην ώρα του καθήκοντος, ως εκ τούτου ο ίδιος ο οργανισμός δεν προσφέρει αυτές τις πληροφορίες.
7. **attack** = το είδος της εκάστοτε επίθεσης.
8. **role** = διευκρινίζεται με βάση τις πληροφορίες από τα αρχικά δεδομένα αν το εκάστοτε θύμα επίθεσης είναι δημοσιογράφος ή εργαζόμενος στα MME
   - Στις περιπτώσεις των θανάτων η Committee to Protect Journalists διαχωρίζει τους δημοσιογράφους από τους εργαζόμενους στα ΜΜΕ, ενώ στις περιπτώσεις των φυλακίσεων και των εξαφανίσεων καταμετρά τα περιστατικά συνολικά.
9. **region** = διευκρινίζεται με βάση τις πληροφορίες από τη στήλη "location" αν η εκάστοτε επίθεση πραγματοποιήθηκε σε κάποια χώρα της Μέσης Ανατολής ή σε κάποια άλλη χώρα στον υπόλοιπο πλανήτη.
    - Ως χώρες της Μέσης Ανατολής θεωρούμε, με βάση την πλειοψηφία των επιστημονικών πηγών, τις εξής: Αφγανιστάν, Μπαχρέιν, Αίγυπτος, Ιράν, Ιράκ, Ισραήλ, Ιορδανία, Κουβέιτ, Λίβανος, Ομάν, Παλαιστίνη, Κατάρ, Σαουδική Αραβία, Συρία, Τουρκία, Ηνωμένα Αραβικά Εμιράτα, Υεμένη.
    - Η Committee to Protect Journalists στην καταμέτρηση των περιστατικών και την ανάλυση των δεδομένων της συμπεριλαμβάνει τα περιστατικά που έλαβαν χώρα στο Ισραήλ και τα περιστατικά που έλαβαν χώρα στα Κατεχόμενα Παλαιστινιακά Εδάφη στην ίδια γεωγραφική περιοχή με το όνομα «Israel and the Occupied Palestinian Territories», επομένως, προς αποφυγή σύγχυσης των δεδομένων, ακολουθήσαμε και εμείς το ίδιο μοτίβο.
10. **year** = διευκρινίζεται με βάση τις πληροφορίες από τη στήλη "date" η χρονιά που πραγματοποιήθηκε η εκάστοτε επίθεση.
11. **staff_freelancer** = διευκρινίζεται με βάση τις πληροφορίες από τη στήλη "organisation" αν ο εκάστοτε δημοσιογράφος ή εργαζόμενος στα ΜΜΕ είναι υπάλληλος σε κάποιο μέσο ή οργανισμό ή αν εργάζεται ως freelancer.
