```python
import pandas as pd

pd.set_option('display.max_rows', None)
```

```python
j_killed = pd.read_csv('j_killed.csv')
```

```python
mw_killed = pd.read_csv('mw_killed.csv')
```

```python
imprisoned = pd.read_csv('imprisoned.csv')
```

```python
missing = pd.read_csv('missing.csv')
```

```python
j_killed['attack'] = 'Killed'
j_killed['role'] = 'Journalist'
mw_killed['attack'] = 'Killed'
mw_killed['role'] = 'Media Worker'
missing['attack'] = 'Missing'
missing['role'] = 'Journalist or Media Worker'
imprisoned['attack'] = 'Imprisoned'
imprisoned['role'] = 'Journalist or Media Worker'
```

```python
df = pd.concat([j_killed, mw_killed, imprisoned, missing])
```

```python
# rename 'Male/Female' value
df['gender'] = df.gender.replace({'male':'Male',
                                  'female':'Female'})
```

```python
df['region'] = ''
```

```python
df['year'] = df.date.str.extract(r',\s+(\d\d\d\d)')
```

```python
middle_east = ['Iraq', 'Iran', 'Egypt', 'Bahrain', 'Israel and the Occupied Palestinian Territory', 'Jordan',
               'Kuwait', 'Saudi Arabia', 'Syria', 'Lebanon', 'Oman', 'Qatar', 'Turkey', 'UAE', 'Yemen']

df['region'] = ''

for i, item in df.iterrows():
    if item['location'] in middle_east:
        df.at[i, 'region'] = 'Middle East'
    else:
        df.at[i, 'region'] = 'All other parts of the world'
```
