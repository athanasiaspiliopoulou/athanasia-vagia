```python
import pandas as pd

pd.set_option('display.max_rows', None)
```

```python
df_journalists_killed = pd.read_csv('journalists_killed.csv')
```

```python
# change Pakistan in Murder
for i, item in df_journalists_killed.iterrows():

    if (item['type_of_death'] == 'Pakistan'):

        df_journalists_killed.at[i, 'type_of_death'] = 'Murder'
```

```python
# check empty values
df_journalists_killed[df_journalists_killed.type_of_death.isnull()==True]
```

```python
# check 'Unknown'
df_journalists_killed[df_journalists_killed.type_of_death=='Unknown']
```

```python
# replace 'Unknown' value with 'Motive Unconfirmed'
for i, item in df_journalists_killed.iterrows():

    if (item['type_of_death'] == 'Unknown'):

        df_journalists_killed.at[i, 'type_of_death'] = 'Motive Unconfirmed'
```

```python
# replace 'NaN' value with 'Motive Unconfirmed'
df_journalists_killed['type_of_death']=df_journalists_killed.type_of_death.fillna('Motive Unconfirmed')
```

```python
# checking the values of Motive Unconfirmed, if Nan and Unkown values have been changed 
df_journalists_killed.type_of_death.value_counts(dropna=False)
```

```python
# rename 'Freelancer' value
df_journalists_killed['organization'] = df_journalists_killed.organization.replace({'Freelance':'Freelancer',
                                  'freelance':'Freelancer'})
```

```python
# replace 'NaN' value with 'Motive Unconfirmed'
df_journalists_killed['organization']=df_journalists_killed.organization.fillna('Unknown')
df_journalists_killed
```

```python
df_journalists_killed.to_csv('j_killed.csv',index=False)
```

```python
df_media_workers_killed = pd.read_csv('media_workers_killed.csv')
```

```python
# replace 'NaN' value with 'Motive Unconfirmed'
df_media_workers_killed['type_of_death']=df_media_workers_killed.type_of_death.fillna('Motive Unconfirmed')
```

```python
# rename 'Freelancer' value
df_media_workers_killed['organization'] = df_media_workers_killed.organization.replace({'Freelance':'Freelancer',
                                  'freelance':'Freelancer'})
```

```python
# correct mistakes
for i, item in df_media_workers_killed.iterrows():

    if (item['location'] == '\tIraq'):

        df_media_workers_killed.at[i, 'location'] = 'Iraq'
```

```python
df_media_workers_killed.to_csv('mw_killed.csv',index=False)
```
