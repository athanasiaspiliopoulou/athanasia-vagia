```python
import pandas as pd

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/athanasiaspiliopoulou/athanasia-vagia/main/PROJECT/df.csv')
```

```python
df.head()
```

```python
df['region'] = ''
for i, item in df.iterrows():
    
    if item['location'] == 'Iraq' or item['location'] == 'Iran' or item['location'] == 'Egypt' or item['location'] == 'Bahrain' or item['location'] == 'Israel and the Occupied Palestinian Territory' or item['location'] == 'Jordan' or item['location'] == 'Kuwait' or item['location'] == 'Lebanon' or item['location'] == 'Oman' or item['location'] == 'Qatar' or item['location'] == 'Saudi Arabia' or item['location'] == 'Syria' or item['location'] == 'Turkey' or item['location'] == 'UAE' or item['location'] == 'Yemen':
        
        df.at[i, 'region'] = 'Middle East'
        
    else:
        
        df.at[i, 'region'] = 'All other countries'
```

```python
df.head()
```

```python
#just checking the classification

pd.DataFrame(df.groupby('region').location.value_counts(dropna=False))
```

```python
df.to_csv('regions_fixed.csv',index=False)
```
