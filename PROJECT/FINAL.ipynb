# Πόσο ελεύθεροι και ασφαλείς είναι άραγε οι δημοσιογράφοι στη Μέση Ανατολή;

#### Τα τελευταία χρόνια τα γεγονότα που επιβεβαιώνουν το έλλειμμα δημοκρατίας και ελευθερίας στις χώρες της Μέσης Ανατολής πληθαίνουν. Σε τέτοιες περιπτώσεις η ελεύθερη, ανεξάρτητη και ασφαλής δημοσιογραφία είναι πιο σημαντική από ποτέ. Τελικά, είναι όμως έτσι στην πράξη;

![blood press vest](blooded-press-vest.jpeg)

*(Arafat Barbakh/Reuters. Ένας δημοσιογράφος κρατάει το ματωμένο δημοσιογραφικό γιλέκο του Παλαιστίνιου δημοσιογράφου Mohammad Soboh, ο οποίος σκοτώθηκε από ισραηλινά πυρά την ώρα που έκανε ρεπορτάζ στη Λωρίδα της Γάζας κατά τη διάρκεια του πολέμου μεταξύ Ισραήλ και Χαμάς.)*

Έχουν περάσει σχεδόν 3 μήνες από την έναρξη του νέου πολέμου μεταξύ Ισραήλ και Χαμάς και οι αριθμοί των θανάτων, των τραυματισμών και των υλικών καταστροφών ολοένα και αυξάνονται. Ανάμεσα σε αυτούς τους αριθμούς βρίσκεται και ο αριθμός των επιθέσεων εναντίον των δημοσιογράφων και εργαζομένων στα ΜΜΕ, ο οποίος σύμφωνα με τη Committee to Protect Journalists (CPJ) «οδήγησε στην **πιο θανατηφόρα περίοδο** για τους δημοσιογράφους από τότε που η CPJ άρχισε να συλλέγει δεδομένα το 1992». Μέχρι τις 4 Ιανουαρίου 2024, οι έρευνες της CPJ έδειξαν ότι **τουλάχιστον 77 δημοσιογράφοι και εργαζόμενοι στα μέσα ενημέρωσης** συγκαταλέγονταν στους περισσότερους από 23.000 νεκρούς από την έναρξη του πολέμου στις 7 Οκτωβρίου.

Είναι όμως μόνο οι επιθέσεις που καταγράφονται σε περίοδο πολέμου που εγείρουν ανησυχίες για την ελευθερία και την ασφάλεια των δημοσιογράφων στη Μέση Ανατολή; Όπως αποδεικνύουν οι έρευνες των Δημοσιογράφων Χωρίς Σύνορα, υπάρχουν και άλλες περιπτώσεις σε χώρες της Μέσης Ανατολής κατά τις οποίες ο Τύπος φιμώνεται με τον έναν ή τον άλλον τρόπο. Έξι χρόνια μετά την απόπειρα πραξικοπήματος, μετά την οποία η Τουρκία έγινε η μεγαλύτερη φυλακή δημοσιογράφων στον κόσμο, **μέσα σε μόλις μισό χρόνο, 30 δημοσιογράφοι** που εργάζονταν στα φιλοκουρδικά μέσα ενημέρωσης συνελήφθησαν και **25 από αυτούς φυλακίστηκαν** (στοιχεία Reporters Without Borders), χωρίς καν να κατηγορηθούν ή να καταδικαστούν. Επίσης, στο Ιράν, έπειτα από το θάνατο της Mahsa Amini στις 16 Σεπτεμβρίου του 2022, ενώ τελούσε υπό κράτηση της αστυνομίας των ηθών, συνολικά **79 δημοσιογράφοι** που προσπάθησαν να ερευνήσουν την υπόθεση, φυλακίστηκαν με την κατηγορία της αντικυβερνητικής προπαγάνδας (στοιχεία Reporters Without Borders).

Αυτές οι περιπτώσεις, όμως, δεν αποτελούν είδηση για τους δημοσιογράφους στη Μέση Ανατολή. Μόνο στον 21ο αιώνα η Μέση Ανατολή μετρά **25 διαφορετικές εμπόλεμες συρράξεις**, πολλές από τις οποίες είναι ακόμα εν εξελίξει, ενώ οι πολιτικές και κοινωνικές αναταραχές θα μπορούσε να πει κανείς πως δεν έχουν ποτέ τελειωμό. Με τις χώρες της Μέσης Ανατολής να βρίσκονται **μεταξύ των θέσεων 97 και 177 στον Παγκόσμιο Δείκτη Ελευθερίας του Τύπου**, πόσο ελεύθεροι και ασφαλείς είναι άραγε οι δημοσιογράφοι στη Μέση Ανατολή;

### Η δημοσιογραφία στη Μέση Ανατολή

Αν λάβει κανείς υπόψιν του ότι οι χώρες της Μέσης Ανατολής (Αφγανιστάν, Μπαχρέιν, Αίγυπτος, Ιράν, Ιράκ, Ισραήλ, Ιορδανία, Κουβέιτ, Λίβανος, Ομάν, Παλαιστίνη, Κατάρ, Σαουδική Αραβία, Συρία, Τουρκία, Ηνωμένα Αραβικά Εμιράτα, Υεμένη) είναι **μόλις 17 χώρες από τα συνολικά 196 κράτη του πλανήτη** κι όμως αυτό το σύνολο των 17 χωρών έναντι των υπολοίπων 179 σε αρκετές χρονιές μόνο κατά τη σύντομη διάρκεια του 21ου αιώνα καταφέρνει να μετρά περισσότερες επιθέσεις κατά του Τύπου από ότι όλος ο υπόλοιπος πλανήτης, με μαθηματική ακρίβεια οδηγείται στο συμπέρασμα ότι οι δημοσιογράφοι στη Μέση Ανατολή καλούνται να αντιμετωπίσουν πολύ πιο επικίνδυνες και ανελεύθερες συνθήκες εργασίας και διαβίωσης και μάλιστα με μεγαλύτερη συχνότητα.

Σαφώς, πρόκειται για μια γεωγραφική περιοχή στην οποία ο ένας πόλεμος διαδέχεται τον άλλον και η μία κοινωνικοπολιτική ένταση απορρέει από την άλλην, με αποτέλεσμα να μιλάμε για χαμηλά ποσοστά δημοκρατίας και δικαιοσύνης. Πιο συγκεκριμένα παρατηρείται ότι οι αριθμοί επιθέσεων στη Μέση Ανατολή αυξάνονται στις περιόδους πολέμων και έντονων αναταραχών. Την περίοδο 2003-2007, όπου παρατηρείται μεγάλη διαφορά με τις υπόλοιπες γεωγραφικές περιοχές, η Μέση Ανατολή ήταν αντιμέτωπη με τον πόλεμο στο Ιράκ (2003-2011), τη σύγκρουση μεταξύ Ιράν και των Κούρδων επαναστατών, τη σύγκρουση μεταξύ Φατάχ και Χαμάς στα Κατεχόμενα Παλαιστινιακά Εδάφη και τη σύγκρουση μεταξύ Ισραήλ και Ιράν. Την περίοδο 2012-2013 συνέβαινε η ανθρωπιστική κρίση στην Υεμένη, ο εμφύλιος πόλεμος στη Συρία, ο οποίος επηρέασε και τον Λίβανο, και ο πόλεμος στο Ιράκ (2013-2017), που συνεχίστηκε και κατά το 2016, όταν παράλληλα συνέβαινε ο εμφύλιος πόλεμος στην Υεμένη και ήταν η χρονιά όπου η προσπάθεια πραξικοπήματος στην Τουρκία άλλαξε τα δεδομένα ελευθερίας του Τύπου στη χώρα.

Ο αριθμός θανάτων των δημοσιογράφων στη Μέση Ανατολή αγγίζει **το 77% των συνολικών επιθέσεων κατά του Τύπου στη Μέση Ανατολή**. Κι ενώ θα μπορούσε να θεωρηθεί ένας λογικός αριθμός λόγω των δεκάδων εμπόλεμων συρράξεων στην περιοχή, σύμφωνα με το άρθρο 8 του Καταστατικού της Ρώμης, **«η στόχευση δημοσιογράφων είναι έγκλημα πολέμου»**.

### Γιατί οι φυλακές στη Μέση Ανατολή γεμίζουν με δημοσιογράφους;

Μεγάλο ποσοστό, όμως, παρουσιάζουν και οι φυλακίσεις των δημοσιογράφων και συχνά οι χώρες της Μέσης Ανατολής θεωρούνται οι μεγαλύτερες φυλακές δημοσιογράφων στον κόσμο, με την Τουρκία και το Ιράν να έχουν την πρωτοκαθεδρία. Μάλιστα, στις περισσότερες περιπτώσεις τα κατηγορητήρια εναντίον των δημοσιογράφων δεν δημοσιεύονται καν από τις αρμόδιες αρχές και οι Δημοσιογράφοι χωρίς Σύνορα, αλλά και η Committee to Protect Journalists καταγγέλλουν **«καταχρηστική σύλληψη δημοσιογράφων»**. Τα αιτήματα και των δύο οργανισμών προς τις αρχές για προστασία των δικαιωμάτων των δημοσιογράφων δεν έχουν λάβει ποτέ καμία απάντηση.

Οι πολυετείς έρευνες των Δημοσιογράφων χωρίς Σύνορα φέρνουν στο φως της δημοσιότητας και τις συνθήκες κράτησης των δημοσιογράφων. Σύμφωνα με τα όσα δηλώνουν -ανώνυμα- οι ίδιοι δημοσιογράφοι που έχουν δεχθεί αντίστοιχες επιθέσεις, κατά τη διάρκεια της κράτησής τους βιώνουν σωματική και ψυχολογική βία, ενώ η δικαστική διαφθορά και η γραφειοκρατική δυσλειτουργία συνδυάζονται για να τους εγκλωβίσουν στους σκοτεινούς διαδρόμους της δικαιοσύνης που χωλαίνει. Οι δημοσιογράφοι μερικές φορές απλώς εξαφανίζονται για ώρες πριν μάθουν τα αγαπημένα τους πρόσωπα ότι τους έβαλαν στην απομόνωση. Όταν ορίζεται ακρόαση, η ημερομηνία μπορεί να ανακοινωθεί και στη συνέχεια να ακυρωθεί, να πραγματοποιηθεί χωρίς να ειδοποιηθούν οι δικηγόροι των δημοσιογράφων ή να ειδοποιηθούν μόνο την τελευταία στιγμή.

«Κανένα ενιαίο σύνολο κανόνων δεν υπαγορεύει τι μπορεί και τι δεν μπορεί να ειπωθεί στα μέσα ενημέρωσης», είπε ένας Ιρανός δημοσιογράφος στους RSF υπό τον όρο της ανωνυμίας. «Εάν υπήρχαν σαφείς κανόνες, θα ξέραμε πώς να τους ξεπεράσουμε. Οι αρχές είναι αρκετά έξυπνες για να το καταλάβουν αυτό. **Όλα είναι αυθαίρετα, επομένως όλα απαγορεύονται**».

Συγκεκριμένα στο Ιράν, οι δημοσιογράφοι φιμώνονται καθώς έχουν μόνιμο τον φόβο τις τιμωρίας. Οι δικαστές συχνά αναστέλλουν τις ποινές για ορισμένο χρονικό διάστημα, διατηρώντας το δικαίωμα να τις εφαρμόσουν εάν ο δημοσιογράφος διαπράξει άλλο «έγκλημα». Ασκούν έτσι μια μορφή μακροχρόνιας λογοκρισίας και επιτήρησης. Ορισμένοι δημοσιογράφοι γλιτώνουν τη φυλάκιση, αλλά συχνά τους απαγορεύεται, συμπεριλαμβανομένων εκείνων που αποφυλακίζονται με εγγύηση, να εγκαταλείψουν τη χώρα, να ασκήσουν δημοσιογραφία και να μιλήσουν στα μέσα ενημέρωσης.

Οι Δημοσιογράφοι χωρίς Σύνορα καταγγέλλουν επίσης και τη μόνιμη απειλή που κρέμεται πάνω από τους δημοσιογράφους που έχουν διαφύγει και ζουν πια στο εξωτερικό. Συχνά δέχονται απειλές από τις αρχές, οι οποίες μερικές φορές φτάνουν στο σημείο να κρατούν αυθαίρετα στενούς συγγενείς που ζουν ακόμα στο Ιράν. Αυτό συνέβη με τη Shima Shahrabi, τη συντάκτρια της περσικής ειδησεογραφικής ιστοσελίδας Iran Wire. Αφού έφυγε από τη χώρα, ο αδελφός της, ο δημοσιογράφος του ραδιοφώνου Sajjad Shahrabi, συνελήφθη στο Ιράν τον Μάιο του 2023, ανακρίθηκε για τη δουλειά της αδερφής του και φυλακίστηκε για περισσότερο από ένα μήνα.

### Ο παράγοντας «φύλο» και οι επιθέσεις κατά του Τύπου

Μπορεί τα δεδομένα να δείχνουν ότι οι άντρες δημοσιογράφοι στη Μέση Ανατολή δέχονται περισσότερες επιθέσεις από ό,τι οι γυναίκες δημοσιογράφοι, όμως τί ποσοστό γυναικών εργάζεται ως δημοσιογράφοι στη Μέση Ανατολή και τί ποσοστό αντρών;

Ο ακριβής αριθμός γυναικών δημοσιογράφων στη Μέση Ανατολή δεν είναι εύκολα προσβάσιμος, όμως με βάση την έρευνα της Παγκόσμιας Τράπεζας για τη συμμετοχή των γυναικών στο εργατικό δυναμικό στη Μέση Ανατολή το 2022, **«οι γυναίκες στη Μέση Ανατολή το 2022 αποτελούσαν μόλις το 19 τοις εκατό του εργατικού δυναμικού** - ένα από τα χαμηλότερα ποσοστά στον κόσμο». Επομένως, **το 12% από τις συνολικά 647 επιθέσεις κατά του Τύπου στη Μέση Ανατολή** δεν αποτελεί ένα μικρό ποσοστό, όταν πρόκειται για μια μικρή πληθυσμιακά επαγγελματική ομάδα, αυτή των γυναικών δημοσιογράφων. Πόσο μάλλον, αν στη συλλογιστική συμπεριληφθεί και το γεγονός ότι αυτό το 12% μετρά και επιθέσεις παλαιότερων δεκαετιών, όταν οι γυναίκες σε αυτές τις χώρες απαγορευόταν ακόμα να δουλεύουν βάσει νόμου σε αντίθεση με τους άντρες, με αποτέλεσμα αυτά τα δύο ποσοστά να φαίνονται ίσα ως προς την ποσοτική τους ανάλυση, αλλά ουσιαστικά να είναι άνισα ως προς την ποιοτική τους ανάλυση.

Τα τελευταία χρόνια οι πιο συχνές επιθέσεις κατά των γυναικών δημοσιογράφων στη Μέση Ανατολή είναι αυτές **των φυλακίσεων με βάση το φύλο τους και το… πόσο καλές μουσουλμάνες είναι ή δεν είναι!** Στο Ιράν σε λιγότερο από ένα χρόνο **τουλάχιστον 30 γυναίκες δημοσιογράφοι φυλακίστηκαν**, είτε γιατί επιχείρησαν να μιλήσουν ανοιχτά για την υπόθεση της Mahsa Amini, είτε γιατί εναντιώθηκαν στους αυστηρά θρησκευτικούς νόμους της αστυνομίας των ηθών, είτε γιατί οι ίδιες δεν τηρούσαν το νόμο περί υποχρεωτικής χρήσης μαντίλας σε δημόσιο χώρο. Ένα από τα πιο χαρακτηριστικά παραδείγματα είναι η υπόθεση της **Nazila Maroufian**, της δημοσιογράφου που έκανε γνωστή τη δολοφονία της Mahsa Amini, η οποία, παρόλο που έχει φυλακιστεί και αποφυλακιστεί δύο φορές για «δημοσίευση προπαγάνδας», κατήγγειλε επώνυμα και δημόσια στους RSF ότι δέχθηκε σεξουαλική επίθεση όταν συνελήφθη.

### Προστατεύονται εν τέλει κάπως οι δημοσιογράφοι στη Μέση Ανατολή;

Κι εκεί που θα πίστευε κανείς ότι το να εργάζεται ένας δημοσιογράφος σε κάποιο μέσο ή κάποιον οργανισμό θα του προσέφερε το δικαίωμα να απολαμβάνει την προστασία που του προσφέρει η φήμη του μέσου ή του οργανισμού, τα δεδομένα αποδεικνύουν πως κάτι τέτοιο στη Μέση Ανατολή δεν παίζει κανέναν ρόλο. Μέσα ενημέρωσης, δημοσιογραφικοί οργανισμοί και πρακτορεία ειδήσεων που έχουν μόνιμους ή απεσταλμένους συνεργάτες δημοσιογράφους στη Μέση Ανατολή, έχει αποδειχθεί πως **απέναντι στους στόχους των κυβερνήσεων δεν έχουν καμία δύναμη, όσο κι αν θέλουν να προστατεύσουν τους συνεργάτες τους**.

Για παράδειγμα, το **Al Jazeera**, ενώ έχει υποβάλει αίτημα στο **Διεθνές Ποινικό Δικαστήριο** να ερευνήσει και να διώξει τους υπεύθυνους για τον πυροβολισμό της δημοσιογράφου **Shireen Abu Akleh** τον Μάιο του 2022 ενώ κάλυπτε μια ισραηλινή στρατιωτική επιδρομή στον προσφυγικό καταυλισμό Τζενίν στην κατεχόμενη Δυτική Όχθη, και έχει παράσχει όλα τα απαιτούμενα αποδεικτικά στοιχεία από τον Δεκέμβριο του 2022, το Διεθνές Ποινικό Δικαστήριο δεν έχει προβεί στις απαιτούμενες ενέργειες. Την ίδια προσπάθεια έχει ξεκινήσει και αυτή την περίοδο το αραβικό μέσο ενημέρωσης για την περίπτωση δολοφονίας του εικονολήπτη του, **Samer Abu Daqqa**, στη Γάζα, όταν ο ισραηλινός στρατός εμπόδισε τα ασθενοφόρα και τις ομάδες διάσωσης να μεταβούν στο σημείο όπου αιμορραγούσε ο Abu Daqqa εξαιτίας των ισραηλινών πυρών εναντίον του ιδίου και του δημοσιογράφου, επίσης εργαζόμενου του Al Jazeera, **Wael Al-Dahdouh**. Παρόλο που είναι ακόμα νωρίς για να γνωρίζουμε την εξέλιξη αυτής της υπόθεσης, οι αντίστοιχες παρελθοντικές προσπάθειες για διαφύλαξη των δικαιωμάτων των δημοσιογράφων στη Μέση Ανατολή έχουν μείνει μετέωρες.

**Ο Sherif Mansour, συντονιστής προγράμματος της CPJ για τη Μέση Ανατολή και τη Βόρεια Αφρική** τονίζει ότι «οι δημοσιογράφοι είναι πολίτες που κάνουν σημαντική δουλειά σε περιόδους κρίσης και **δεν πρέπει να γίνονται στόχος αντιμαχόμενων πλευρών**». Αντίστοιχες δηλώσεις κάνει κατά καιρούς και ο ΟΗΕ, ο οποίος έχει δημιουργήσει από το 2012 και το **Σχέδιο Δράσης του ΟΗΕ για την ασφάλεια των δημοσιογράφων**, όμως οι αριθμοί δείχνουν πως όλες αυτές οι δηλώσεις εν τέλει δεν αρκούν για να είναι ελεύθεροι και ασφαλείς οι δημοσιογράφοι στη Μέση Ανατολή.
